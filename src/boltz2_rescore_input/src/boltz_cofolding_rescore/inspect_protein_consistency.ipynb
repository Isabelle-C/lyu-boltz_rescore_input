```python
import os
from boltz.rescore.load.protein import Protein
```

```python
all_pdb_path = "/lustre/fs6/lyu_lab/scratch/ichen/data/boltz_runs/ampc_pdb"
```

```python
def get_protein_coords_dict(protein_coords):
    protein_coords_dict = {}
    for i in protein_coords:
        protein_coords_dict[(i[0], i[1], i[2])] = [i[3], i[4], i[5]]
    return protein_coords_dict
```

```python
def compare_coord_dicts(dict1, dict2):
    # Find common keys
    common_keys = set(dict1.keys()) & set(dict2.keys())
    # Compare values for common keys
    mismatches = []
    for key in common_keys:
        if dict1[key] != dict2[key]:
            mismatches.append((key, dict1[key], dict2[key]))
    print(f"Total common keys: {len(common_keys)}")
    print(f"Number of mismatches: {len(mismatches)}")
    return mismatches
```

```python
protein_list = os.listdir(all_pdb_path)
```

```python
protein_list[0]
```

```python
first_pdb_path = f"{all_pdb_path}/{protein_list[0]}"


protein = Protein(first_pdb_path)
protein_seq = protein.get_sequence()
protein_coords = protein.get_coords()
```

```python
first_pdb_coords = get_protein_coords_dict(protein_coords)
```

```python
for i in protein_list[1:]:
    pdb_path = f"{all_pdb_path}/{i}"
    print(i)
    protein = Protein(pdb_path)
    protein_seq = protein.get_sequence()
    protein_coords = protein.get_coords()

    current_pdb_coords = get_protein_coords_dict(protein_coords)

    mismatches = compare_coord_dicts(first_pdb_coords, current_pdb_coords)
    
    if len(mismatches)>0:
        raise ValueError(f"Mismatch found in {i} compared to the first PDB file {protein_list[0]}:\n{mismatches}")
```
